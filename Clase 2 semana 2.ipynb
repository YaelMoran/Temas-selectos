un **diccionario** es una estructura de datos que permite guardar informacion en pares: clave -->valor.
* a diferencia de las listas (que usan posiciones), los diccionares usan nombres (**claves**) para acceder a los datos
* se usan llaves {}

```python
reactor = {
    'tipo': 'CSTR',
    'volumen': 5.0,
    'temperatura': 350,
    'presion': 2.5,
}
reactor
```

```python
#ejemplo de diccionario
cafe = {
    'leche': 'deslactosada light',
    'tamaño' : ' chico',
    'tipo' : 'descafeinado',
    'nombre': 'yael'
}
print(cafe)
```

Se puede acceder a los elementos, insertarlos o formar conjuntos con ellos utilizando la misma sintaxis para acceder a los elementos de una lista.

```python
print(cafe['tamaño'])
```

```python
cafe['tamaño'] = 'grande' #cambiar un elemento en este caso el de tamaño
cafe['azúcar'] = 'mascabado' #para agregar un nuevo elemento
```

```python
print(cafe)
```

**Funciones en python**
Las funciones son el metodo mas importante en python para organizr, reutilizar y simplificar el codigo

Regla general:

Si un mismo código se va a usar mas de una vez, conviene escribir una funcion.

En python se declaran con la palabra clave **def**, contiene un bloque de codigo, con uso opcional de la palabra clave **return**

```python
def sumar (a, b): #lo que se encuentra dentro de los parentesis son los parametros de entrada
  x = a + b
  return x
```

¿Como usar la funcion?

```python
sumar(15,75)
```

```python
def prom (a,b,c,d):
  prom = (a+b+c+d)/4
  return prom
```

```python
prom(45,12,546,89)
```
