**Ejercicio 1. Ley de los gases ideales**

La ecuación de los gases ideales relaciona la presión, el volumen, la cantidad de sustancia y la temperatura mediante la expresión:

#                                  PV = nRT


* Datos:

* P = 1,2 atm

* R =?

* T = 273 K

* V = 0,7 L

Realice una busqueda para obtener el valor de la constante de los gases (R) en las unidades requeridas. Determina el valor de los moles n del sistema.

```python
#Primero colocamos las variables del sistema
P = 1.2 #atm
T = 273 #K
V = 0.7 #L
#Realizamos la busqueda para obtener el valor de R que necesitamos
R = .082 #Latm/molK

n = (P*V)/(R*T) #Ponemos la ecuación de los gases ideales con las variables ya definidas
print(f'El valor de los moles es {n:.5f} mol') #imprimimos la cantidad de moles llamando a la variable n con 5 decimales

```

* A = 5 × 10^7 s^−1

* Ea = 75,000 J·mol^−1

* R = 8,314 J·mol^−1·K^−1

* T = 350 K

Calcula la constante de velocidad k.

```python

#Definimos las variables dadas por el sistema
A = 5*10**7
Ea = 75000
R = 8.314
T = 350
e = math.exp #para llamar a la funcion exponencial usamos math.exp

#podemos usar directamente la funcion math.exp o definirla como lo hice yo

k = A*e(-Ea/(R*T)) #escribimos la ecuacion de la cte de velocidad
print(f'La constante de velocidad es {k:.8f} s^-1') #imprimimos la función dando 8 decimales y con las unidades
```

Dada una ecuación cuadrática

# ax2 + bx + c = 0,
El discriminante (∆ = b − 4ac) determina el tipo de raíces:

• ∆ > 0 → Dos raíces reales distintas

• ∆ = 0 → Una raíz real doble

• ∆ < 0 → Dos raíces complejas conjugadas

Realice el código que indique el tipo de raíces e emprima el valor si son reales.

```python
#colocamos las variables con input para poder poner distintos valores para a, b, c
a = int(input('Introduce el valor de a: ')) #al usar int solo permitimos la entrada de números enteros
b = int(input('Introduce el valor de b: '))
c = int(input('Introduce el valor de c: '))

delta = (b**2) - (4*a*c)

if delta > 0:
  x1 = (-b + math.sqrt(delta))/(2*a) #usamos la funcion math.sqrt importada de la biblioteca de math, para hacer las raíces cuadradas
  x2 = (-b - math.sqrt(delta))/(2*a)
  1print('Dos raíces reales distintas:', x1, x2) #si se cumple, pedimos que muestre los valores de las raíces

elif delta == 0:
  x = -b/(2*a)
  print('Una raíz real doble:',x) #pedimos que si se cumple nos de el valor de la raíz real doble

else:
  print('Raíces complejas conjugadas')
```

**Ejercicio 5. Número de Reynolds (flujo de fluidos)**

El número de Reynolds (Re) se utiliza para caracterizar el régimen de flujo (laminar, de transición o turbulento):

Re = rhovD/µ

Datos:

rho = 1000 kg·m^−3

v = 1.2 m·s^−1

D = 0.05 m

µ = 0.001 Pa·s

Cálcula el valor del número de Reynolds.

```python
#primero definimos las variables que vamos a usar
rho = 1000
v = 1.2
D = .05
µ = .001

Re = rho*v*D/µ #Ponemos la ecuación del número de Reynolds con las variables ya definidas
print(f'El número de Reynolds es: {Re:.2f}') #imprimimos la variable Re, agregando el tipo de datos float(f) para que nos de con decimal si es que llegara a haber, y usamos lo que esta entre llaves para que nos de 2 decimales.
```

**Ejercicio 6. Uso de condicionales: Determinación del régimen de flujo.**

Aplicar estructuras de control if, elif y else para determinar el tipo de flujo a partir del número de Reynolds calculado.

El programa deberá solicitar al usuario los valores necesarios mediante la función input(),correspondientes a:
* Densidad del fluido, rho [kg/m3]
* Velocidad promedio del fluido, v [m/s]
* Diámetro interno de la tubería, D [m]
* Viscosidad dinámica, mu
 [Pa·s]

Con los valores introducidos, el código deberá calcular el número de Reynolds:

Re = rhovD/mu

Posteriormente, el programa determinará el régimen de flujo de acuerdo con los siguientes criterios:

Re < 2300 → Flujo laminar

2300 ≤ Re ≤ 4000 → Flujo de transición

Re > 4000 → Flujo turbulento


```python
rho = float(input('Introduzca la densidad del fluido en kg/m3: ')) #Denfinimos nuestras variables. Usamos float para permitir números decimales y colocamos las unidades en las cuales debe ingresarse los datos.
v = float(input('Introduzca la velocidad promedio del fluido en m/s: '))
D = float(input('Introduzca el diámetro interno de la tubería en m: '))
mu = float(input('Introduzca la viscosidad dinámica del fluido en Pa/s: '))

Re = rho*v*D/mu #Definimos la ecuacion de Reynolds

if Re < 2300: #Usamos el condicional if para el flujo laminar, pongo lo que esta entre {} para que se muestre el número de Reynolds y diga que tipo es
  print(f'El flujo es de {Re:.2f}, y es laminar')
elif Re >= 2300 and Re <= 400: #Se usa el operador relacional and, porque tenemos que cumplir con 2 condiciones.
  print(f'El flujo es de {Re:.2f} y es de transición')
else:
  print(f'El flujo es de {Re:.2f} turbulento')
```

**Ejercicio 7. Estado del agua respecto al punto crítico**

Aplicar estructuras condicionales para determinar el estado del agua en función de su temperatura y presión, comparándolas con las condiciones críticas.

El punto crítico del agua se define por una temperatura y una presión a las cuales la fase líquida y la fase vapor se vuelven indistinguibles. Los valores son:

Tc = 647,1 K, --------      Pc = 22,06 MPa

El programa deberá solicitar al usuario la temperatura y presión del sistema mediante la función input(), y posteriormente determinar el estado del agua utilizando estructuras condicionales if, elif y else de acuerdo con los siguientes criterios:

* Si T < Tc y P < Pc: el agua se encuentra en fase líquida o vapor.

* Si T > Tc y P > Pc: el agua está en estado supercrítico.

* Si T < Tc y P > Pc: el agua se encuentra en condiciones de líquido comprimido.

* En cualquier otro caso: las condiciones son cercanas al punto crítico.


```python
#Definimos las variables que nos da el problema
Tc = 647 #K
Pc = 22.06 #MPa

T = float(input('Introduce la temperatura del sistema en K: ')) #colocamos el input para ingresar los valores en las unidades que correspondan
P = float(input('Introduce la presion del sistema en MPa: ')) # usamos float para permitir el uso de números decimales

if T < Tc and P < Pc: #Colocamos operador relacional porque se tienen que cumplir con 2 parametros
  print('El agua se encuentra en fase líquida o vapor.')
elif T > Tc and P > Pc:
  print('El agua está en estado supercrítico.')
elif T < Tc and P > Pc:
  print('El agua se encuentra en condiciones de líquido comprimido.')
else:
  print('Las condiciones son cercanas al punto crítico.')
```
