**Ley de los gases ideales**

PV = nRT

Datos de los siguientes valores obtenga la presion del gas.

n = 2.0 mol

R = 0.0821 L*atm/mol*K

T = 298K

V = 10L


```python
n = 2
R = 0.0821
T = 298
V = 10

P = (n*R*T)/V

print(f'Presión del gas: {P:.2f} atm') #Podemros pedir directamente aqui usando la función 'f', para que reconozca puntos decimales. Usar .2f es para que solo nos de 2 decimales despues del punto
```

**Ecuacion de Antoine**

```python
A, B, C = 8.07131, 1730.63, 233.42
T = 80
#Operación
P = 10**(A-B/(T+C))
print(f'Presión de vapor del agua a {T}°C es igual {P:.2f} atm')
```

El numero de Reynolds Re permite saber si el flujo es laminar, de transición o turbulento.
Re=pvD/ mu

```python
rho = 1000
v = 1.2
D = .05
mu = .001

Re=(rho*v*D)/mu
print(f'El numero de Reynolds es {Re:.2f}')
```

**Control de flujo.**

if, elif y else.

Permiten que el proframa tome condiciones segun condiciones. la sentencia **if** es uno de los tipos de secuencia de control de flujo mas conocidos. Comprueba una condicion, se ejecuta si la condicion es verdadera.

```python
x = -56

if x<0:
  print('x es negativo')
```

**if** se ejecuta si la condicion es verdadera

**elif** se ejecuta si la anterior fue falsa, pero esta es verdadera.

**else** se ejecuta si ninguna condicion anterior fue verdadera. No lleva condición.

Una sentencia ***if*** puede ir seguida por uno o mas bloques de ***elif*** y un bloque multifuncional ***else*** si todas las condiciones son *False*

```python
x = 18
if x<0:
  print ('x es negativo') #sangría de 4 espacios
elif x==0:
  print('x es igual a cero')
else:
      print('x es positivo')
```

Ejemplo con input

```python
n = int (input ('Ingresa tu número: '))

if n > 0:
  print ("El número es positivo")
elif n < 0:
  print ("El número es negativo")
else:
  print ("El número es cero")
```

Dada una ecuación cuadrática.

ax^2 + bx + c= 0,

El discriminante (Δ=b^2-4ac) determina el tipo de raíces:

Δ>0, dos raíces reales distintas

Δ = 0 una raíz real doble

Δ < 0 dos raíces complejas conjugadas

Ejercicio: Escribe un código que indique el tipo de raíces e imprima el valor si son reales.

```python
a = int(input('Introduce un valor para a ')) #PARA INSERTAR VALORES DE a, b y c
b = int(input('Introduce un valor para b '))
c = int(input('Introduce un valor para c '))

delta = (b**2)-(4*a*c)
print(f'El determinante es {delta:.2f}')

if delta > 0:
  print('Hay dos raíces reales distintas')
elif delta == 0:
  print('Hay una raíz real doble')
else:
  print('Hay dos raíces complejas conjugadas')
```
