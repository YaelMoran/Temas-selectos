Las listas en Python tienen las siguientes propiedades:


*   Son ordenadas: cada elemento dentro de la lista esta indexado y mantiene el orden definido.
*   Las listas pueden anidar, es decir, contener una lista dentro de la otra.
*   Son mutables: podemos editar, eliminar o agregar elementos en una lista existente (.append(), .remove(),lista 1 [x,y] = y)


```python
#en lugar de decir igual se dice asignación

enteros = [25, 35, 45, 15] #son numeros enteros
cadena =  ["árbol", "quico", "pasto", "flores"] #son palabras DEBEN IR ENTRE COMILLAS
combinado = [75, 25.63, "maquina", "azul" ] #se conbinan enteros, de cadena, y flotantes

lista_pro = [enteros, 17, 75, 157] #para llamar una lista se coloca el nombre de la lista
```

```python
#Para mostrar se puede usar la palabra print o solo colocar el nombre
print(enteros) #muestra el resultado solamente
cadena #comunmente  muestra los calculos y el resultado
```

```python
print (enteros, cadena, combinado, lista_pro) # con print muestra todos los valores
```
