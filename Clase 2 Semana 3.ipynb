**Expresion regular (regex)**

una **expreion regular** es una forma de describir un patron de texto. En vez de buscar una palabra exacta, le dices a Python 'Busca algo que se parezca a esto'

```python
capitales = ['#Ciudad Victo*ria', 'cuerna#vaca', 'Morelia!', 'xalapa', 'Tepic', 'hermos*illo',
 'toluca?', 'San? Luis Potosí', '*Mexicali', 'Villahermosa', 'Tuxtla Gutiérrez#',
 'Colima', 'la Paz#', 'Guadalajara', 'C*ampeche!', 'Durango', 'Zacatecas',
 '!Chetumal', 'Ciudad# de México', 'chilpancingo*', 'Chihua#hua', '?Aguascalientes',
 'Oaxaca de Juárez*', '?Puebla', 'Quer?étaro', 'Mérida', 'guanajuato',
 'pachu*ca', '#Saltillo#', 'Culiacán!', 'Tlaxcala', 'monterrey']
```

**Limpiar una lista**

COn regex puedes:
* buscar patrones en texto (encontrar)
* extraer partes de un teto
* etc

Ejemplo:

a partir de ...


¿Que hacer?
1) Convierte a string

2) quita espacios a inicio y al final

3) elimina simbolos no deseados:

4) normaliza espacion internos

5) capitaliza palabra por palabra con control

```python
import re
def limpiar_cadenas(strings):
  result = []
  for value in strings:
    value = value.strip() #suprime blancos u otro carácter especifico del final
                          #o del principio de una expresion
    result.append(value)
  return result
```

```python
capitales_limpias = limpiar_cadenas(capitales)
capitales_limpias.sort()
```

```python
print(capitales_limpias)
```

```python
capitales_limpias_pro = limpiar_cadenas(capitales_limpias)
capitales_limpias_pro.sort()
```

```python
import re
def limpiar_cadenas(strings):
  result = []
  for value in strings:
    value = value.strip() #suprime blancos u otro carácter especifico del final
                          #o del principio de una expresion
    value = re.sub('[!*#?]', '',  value) #reemplaza todas las ocurrencias de un patrón por el texto que especificamos
    result.append(value)
  return result
```

```python
print(capitales_limpias_pro)
```

```python
capitales_limpias_pro2 = limpiar_cadenas(capitales_limpias)
capitales_limpias_pro2.sort()
```
