CICLO WHILE, TRUE

Sirve para crar un bucle iinfinito que se repetira continuamente hasta que se detenga explicitamente mediante una instruccion como *break*

Se utiliza cuando no se conoce cuantas veces se debe repetir el bucle y se quiere que el usuario decida cuando salir

```python
#Sintaxis
while True:
  n = int(input('Ingresa un número posiivo: '))
  if n >0:
    break #se sale del bucle cuando es una condicion valida
  print('Número no válido. ')
```

```python
password = 'mexico' #definimos una contraseña
while True: #abrimos el ciclo
  contraseña = input('Ingresa la contraseña: ') #para introducir una contraseña
  if password == contraseña: #si la contraseña que ingresamos es igual a la que esta guardada
    print('Contraseña validada') #imprimimos que es correcto, si no es la misma se empieza a repetir el ciclo
    break  #si es correcto entonces termina el ciclo

```

**EJEMPLO**: Teorema de Chebyshev y Regla empirica

```python
lista = [26.1, 26.0, 14.5, 29.3, 19.7,
         22.1, 21.2, 26.6, 31.9,25.0,
         15.9, 20.8, 20.2, 17.8, 13.3,
         25.6, 26.5, 15.7, 22.1, 13.8,
         29.0, 21.3, 23.5, 22.1, 10.2]
```

```python
import statistics as st
```

```python
print('La media es: ', st.mean(lista))
print('La mediana es: ', st.median(lista))
#si no hay una moda clara, puede lanzar StatisticsError
print('La moda es: ', st.mode(lista))
```

Medidas de desvacion
* VARIANZA
* DESV ESTANDAR
* RANGO

```python
#Variazna y desviación estándar muestrales
print('Varianza (muestral):', st.variance(lista))

#poblacionales si fuera TODA la poblacion
print('Varianza (poblacional):', st.pvariance(lista))

#afregar klo k falta
```

```python
#n=4 devuelve tres corets: Q1, Q2, Q3
q1, q2, q3 = st.quantiles(lista, n=4, method='inclusive')
iqr = q3 - q1

print('Q1, Q2, Q3:', q1, q2, q3)
print('IQR:', iqr)
```

**deteccion de outliers con IQR**

```python
li = q1 - 1.5 * iqr
ls = q3 + 1.5 * iqr
outliers = [x for x in lista if x < li or x > ls]

print('Limite inferior:', li)
print('Limite superior:', ls)
print('Outliers:', outliers)
```

**COEFICIENTE DE VARIACION**


```python
cv = st.stdev(lista) / st.mean(lista) * 100
print('CV (%):', cv)
```
