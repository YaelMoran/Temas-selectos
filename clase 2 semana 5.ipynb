```python
import pandas as pd
```

```python
ruta = 'https://raw.githubusercontent.com/marinlopabi/Temas_Selectos_PQ_26-I/refs/heads/main/ARCHIVOS/dataset_distill.csv'
df = pd.read_csv (ruta, sep=';') #cuando es csv se tiene que poner el separador sep=';'
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_") #para quitar los espacios y agregar en su lugar el guión bajo
df.head()
```

Estadistica descriptiva básica

```python
df.describe() #para variables numericas
```

```python
df.dtypes # para saber que tipo de dato es mi columna
```

```python
df['L'] = pd.to_numeric(df['L'], errors='coerce')
df['V'] = pd.to_numeric(df['V'], errors='coerce') #para pasar los valores que no son numeros a valores númericos, el coerce funciona para forzar que se vuelvan números
```

```python
df.describe()
```

```python
pd.set_option('display.max_columns', None) #para que muestre todas las columnas
df.describe()
```

```python
df.describe().style.format('{:.3f}') #para cambiar el formato y ahora nos de 4 decimales solamente
```

```python
tem = df[['T1','T2','T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9','T10', 'T11', 'T12', 'T13', 'T14']]
tem.head()
```

**ASIMETRIA DE FISHER**

si  y1 >0 positiva

si y<1 es neg

```python
asimetria = tem.skew()
print('\nAsimetría de Fisher: \n')
print(asimetria)
```

**CURTOSIS (KURTOSIS)**
el grado de las colas

```python
curtosis = tem.kurt()

print('\nCurtosis de Fisher:\n')
print(curtosis)
```

```python
import matplotlib.pyplot as plt
```

```python
import math
bins = math.sqrt(len(curtosis))
print(bins)
```

```python
tem.hist(bins=10, figsize=(14,10), color = 'aliceblue', edgecolor = 'blue')
plt.tight_layout()
plt.show()
```

mapa de colores https://python-charts.com/es/colores/

```python
columnas = tem.columns

plt.figure(figsize=(15,16))

for i, col in enumerate(columnas,1):
  plt.subplot(4,4,i) #4 filas, 4 columnas de gráficos
  plt.hist(tem[col], bins=20, color='skyblue', edgecolor='black')
  plt.xlabel(col)
  plt.ylabel('Frecuencia')

#plt.sptitle('Distribución de temperaturas y su curtosis (Fisher), fontsize=10')
plt.savefig('histogramas.png', dpi=300,bbox_inches='tight')
plt.show()
```

```python
cols_num = tem.select_dtypes(include='number')
cols_num.plot(kind='box', figsize=(14,6), grid=True, color='purple')
plt.ylabel('temp')
plt.xticks(rotation=100, ha='right')
plt.tight_layout()
plt.show()
```
