**test de mann-kendall**

Se utiliza para detectar si existe una tendencia monótona (creciente o decreciente) en una serie temporal

si es 1 es positivo

```python
import pymannkendall as mk #IMPORTAR LA LIBERIA PYMANKENDALL
import pandas as pd
```

```python
ruta = 'https://raw.githubusercontent.com/marinlopabi/Temas_Selectos_PQ_26-I/refs/heads/main/ARCHIVOS/rama_2023.csv'
df = pd.read_csv (ruta, sep=',') #cuando no sale bien hay que cambiar el separador ()
df.head()
```

```python
df = df.set_index('fecha')
df.head()
```

```python
resultado = mk.original_test(df['PM10'], alpha = .05) #aplicar el metodo mann-kendall, alpha es la covarianza
resultado
```

```python
resultado.slope
```

✈


```python
resultado = mk.original_test(df['PM10'], alpha = .05)

print('=' * 40) #linea de separacion formada por 40 signos
print('Resultado test de mann-kendall ✈') #titulo
print('=' * 40)

print(f'\nTendencia:         {resultado.trend}')
print(f"Significativa:     {'Si' if resultado.h else 'No'}") #resultado.h es True si se rechaza la hipótesis nula y False si no se rechaza

print(f"\n Estadísticos: ")
print(f"  Valor p:     {resultado.p:.4f}") #si p<.05, se considera que la tendencia es significativa
print(f"  Z:          {resultado.z:.4f}") #indica la direccion e intensidad de la tendencia
print(f"  Tau:        {resultado.Tau:.4f}") #mide la asociación monotonica
#positivo = tendencia creciente, negativo = tendencia decreciente
print(f"  S           {resultado.s}") #Direccion --> positiva = creciente. negativa = decreciente

print(f"\n Tendencia de Sen:")
print(f"  Pendiente:   {resultado.slope:.4f}") #cuanto cambia la variable con respecto al tiempo
print(f"  Intercepto:  {resultado.intercept:.4f}") #intercepcion de la recta estimada con la pendiente de sen

print("\n" + '-.'*30)
conclusion= 'Existe una tendencia estadísticamente significativa.' if resultado.h else 'No existe una tendencia estadísticamente significativa.'
print(f'  {conclusion}') #cpnclusion automatica según el valor de resultado.h
#si h es True, existe tendencia significativa
#si h es false, no existe tendencia significativa
print('-.'*30)

```
