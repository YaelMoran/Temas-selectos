**GRÁFICOS UTILES EN ESTADISTICA**

PAQUETE Matplotlib

Es utlizada para crear graficos estaticos, animados e interactivos de manera sencilla y personalizable. Esta diseñada para trabajar de forma similar a MATLAB y se integra facilmente como NumPy y pandas

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(lista, bins=5) #bins son los contenedores
plt.title("histogama")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.show()
```

```python
import math
len(lista)
```

```python
#para saber el numero de bins se tiene que sacar la raíz de el numero de datos
bins = math.sqrt(len(lista))
print(bins)
```

```python
plt.hist(lista, bins=5, color="white", edgecolor='green', alpha=0.8)
plt.title("histogama")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.show()
```

```python
media = np.mean(lista) #media
mediana = np.median(lista) #mediana

plt.hist(lista, bins=5, color="white", edgecolor='green', alpha=.5)
plt.axvline(media, color='red', linewidth=2, label = f"Media = {media:.2f}")
plt.axvline(mediana, color='blue', linewidth=2, linestyle="--", label=f'mediana = {mediana:.2f}')
plt.title("Histograma con media y mediana")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.legend()
plt.show()
```

Para guardar la figura creada

Boxplot / gráfico de cajas


```python
plt.boxplot(lista)
plt.title("Gráfico de cajas")
plt.xlabel("Valor")
plt.show()
```

```python
plt.boxplot(lista, vert=False)
plt.title("Gráfico de cajas")
plt.xlabel("Valor")
plt.show()
```

```python
plt.boxplot(lista,
            vert=False,
            patch_artist=True,
            boxprops=dict(facecolor='skyblue'),
            medianprops=dict(color='green', linewidth=3))
plt.title("Gráfico de cajas")
plt.xlabel("Valor")
plt.show()
```

```python
plt.hist(df['PM10'], bins=55, color = 'gold', edgecolor='black', alpha=0.9)
plt.title("histograma del PM10")
plt.xlabel("valor de PM10")
plt.ylabel("frecuencia")
plt.grid(True)
plt.show()
```

La gráfica esta cargada hacia la izquierda, los ultimos valores de PM10 tienen muy poca frecuencia, es considerablemente

```python
plt.boxplot(df['PM10'],
            vert=False,
            patch_artist=True,
            boxprops=dict(facecolor='skyblue'),
            medianprops=dict(color='red', linewidth=3))
plt.title("Gráfico de PM10")
plt.xlabel("Valor de PM10")
plt.show()
```

mis datos tienen asimetria, se ve mas cargado del lado izquierdo que del lado derecho, ademas de que hay varios valores atípicos que se encuentra en la derecha
