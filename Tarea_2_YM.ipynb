```python
L1 = [7.2, 7.8, 6.8, 8.0, None, '8.2', 5.6, 8.2, 7.7, 7.5, None, 5.8]
L2 = ['6.8', None, 6.8, 6.1, 7.9, 9.4, None]
L3 = [8.5, '9.0', None, 7.7]

```

```python
def limpiar_lista(lista): #Definimos la funcin que ser para una lista
    nueva_lista= [] #creamos una nueva lista que tendra los valores limpios de la lista
    for x in lista: #para cada x en la lista
        if x is not None: #si x no es none, si el valor es none se ignora este punto
            nueva_lista.append(float(x)) #si es no es none, se pasan los valores a flotante
    return nueva_lista #regresa la lista limpia sin None ni cadenas
```

```python
L1_limpia = limpiar_lista(L1) #definimos la nueva variable como L1_limpia llamando la funcin creada en el bloque anterior y agregamos la lista 1
print(L1_limpia) #imprimimos la lista limpia
```

```python
L2_limpia = limpiar_lista(L2) #definimos la nueva variable como L2_limpia llamando la funcin creada
print(L2_limpia) #imprimimos la lista
```

```python
L3_limpia = limpiar_lista(L3) #definimos la nueva variable como L3_limpia llamando la funcin creada
print(L3_limpia)
```

```python
#Ahora sumamos las listas para crear una nueva con todos los valores
tres_listas = L1_limpia + L2_limpia + L3_limpia
tres_listas
```

```python
for i in range(len(tres_listas)): #para cada i en cada elemento de la lista
 for j in range(len(tres_listas) - 1): #para cada j en la longitud de la lista -1, es para que vaya recorriendo elemento por elemento
  if tres_listas[j] > tres_listas[j+1]: #si el valor actual de la lista, es mayor que el siguiente
      tres_listas[j], tres_listas[j+1] = tres_listas[j+1], tres_listas[j] #entonces se intercambiaran los valores para que qiede en orden

print(tres_listas)
```

```python
#si no se hiciera por bucle se podria hacer de la siguiente manera
tres_listas.sort()#usamos .sort para hacerlo en ordenn ascendente
print(tres_listas)
```

```python
cantidad = len(tres_listas) #ahora para saber la cantidad de datos
cantidad_listas = cantidad // 3 #dividimos entre 3 la cantidad de elementos para saber cuanto es para cada lista de manera equitativa
print('La cantidad de elementos para cada lista es: ', cantidad_listas) #mostramos la cantidad
```

```python
lista1_nueva = tres_listas[0:cantidad_listas] #llamamos la nueva lista 1 poniendo entre corchetes la posicion 0 hasta la cantidad de elementos que son por lista
print('La lista 1 nueva es: ', lista1_nueva)

lista2_nueva = tres_listas[cantidad_listas:cantidad_listas*2] #llamamos la nueva lista 2, usando ahora *2 para usar los siguientes 6 datos
print('La lista 2 nueva es: ', lista2_nueva)

lista3_nueva = tres_listas[cantidad_listas*2:cantidad_listas*3] #llamamos la nueva lista 3, ahora iniciando en *2 y termina en *3, debido a la cantidad de datos
print('La lista 3 nueva es: ', lista3_nueva) #ya con esto tenemos las 3 listas en orden ascendente y divididas equitativamente
```

```python
import numpy as np #importamos la libreria

print("Introduce los 9 elementos de la matriz 3x3:") #Aqui pedimos para que se ingrese los datos de la matriz

matriz = [] #la matriz
for i in range(3): #oara un dato i en un rango de 3
    fila = [] #la fila de la matriz
    for j in range(3): #para j en run rango de 3
        valor = float(input(f"Elemento [{i+1},{j+1}]: ")) #aqui es donde el valor que se introduce puede ser flotante, se introducen los elementos y se repite hasta completar el rango previamente establecido
        fila.append(valor)
    matriz.append(fila)

#para calcular el determinante por la regla de Sarrus
det_sarrus = 0


det_sarrus = (matriz[0][0]*matriz[1][1]*matriz[2][2] +
              matriz[0][1]*matriz[1][2]*matriz[2][0] +
              matriz[0][2]*matriz[1][0]*matriz[2][1] -
              matriz[0][2]*matriz[1][1]*matriz[2][0] -
              matriz[0][0]*matriz[1][2]*matriz[2][1] -
              matriz[0][1]*matriz[1][0]*matriz[2][2]) #agragamos la regla de Sarrus

print("\nMatriz:") #para poner este subtitulo ponemos \n para agregar un espacio
for fila in matriz:
    print(fila)

print("\nDeterminante (Regla de Sarrus):", det_sarrus) #aqui acomodamos para que muestre el texto del determinante con la regla de Sarrus

#para calcular con Numpy
matriz_np = np.array(matriz) #colocamos el array para la matriz
det_numpy = np.linalg.det(matriz_np) #usamos la funcion para el determinante con la biblioteca de numpy
print("Determinante (NumPy):", det_numpy) #acomodamos para mostrar el determinante


5
```

```python
#segunda parte
n = int(input("\nCuntos elementos tendrn las listas?: ")) #introducimos un input para pedir el numero de elementos por lista

lista1 = [] #agregamos para lista 1 y lista 2
lista2 = []

print("\nIntroduce los elementos de la primera lista:") #ahora colocamos los elementos dependiendo del numero de elementos definidos por n
for i in range(n): #para i en un rango de numeros definidos por n
    lista1.append(float(input(f"Elemento {i+1}: "))) #agregamos para que se vaya recorriendo en la lista

print("\nIntroduce los elementos de la segunda lista:")
for i in range(n):
    lista2.append(float(input(f"Elemento {i+1}: ")))

producto_escalar = 0
for i in range(n):
    producto_escalar += lista1[i] * lista2[i] #el producto escalar con la suma de los productos de cada elemento de la lista

print("\nLista 1:", lista1) #mostramos la lista 1
print("Lista 2:", lista2) #mostramos la lista 2
print("Producto escalar:", producto_escalar) #mostramos el producto escalar
```

```python
#primero ingresamos listas con los estados y capitales de la respublica mexicana en orden alfabetico
estados = [
    "aguascalientes", "baja california", "baja california sur", "campeche",
    "chiapas", "chihuahua", "coahuila", "colima", "durango", "guanajuato",
    "guerrero", "hidalgo", "jalisco", "mxico", "michoacn", "morelos",
    "nayarit", "nuevo len", "oaxaca", "puebla", "quertaro", "quintana roo",
    "san luis potos", "sinaloa", "sonora", "tabasco", "tamaulipas",
    "tlaxcala", "veracruz", "yucatn", "zacatecas"
]

capitales = [
    "aguascalientes", "mexicali", "la paz", "campeche", "tuxtla gutirrez",
    "chihuahua", "saltillo", "colima", "victoria de durango", "guanajuato",
    "chilpancingo", "pachuca", "guadalajara", "toluca", "morelia",
    "cuernavaca", "tepc", "monterrey", "oaxaca", "puebla", "quertaro",
    "chetumal", "san luis potos", "culiacn", "hermosillo", "villahermosa",
    "ciudad victoria", "tlaxcala", "xalapa", "mrida", "zacatecas"
]

texto = input("Ingresa un en minsculas y con acento: ").lower() #pedimos que se ingrese un texto aadiendo la funcion .lower() para transformar el texto a minusculas para que pueda detectarse en caso de que no cumplan con lo pedido

si_estado = False #si no es estado
si_capital = False #si no es capital

for estado in estados: #definimos si un estado esta en la lista de estados
    if texto == estado: #si el texto ingresado es estado
        si_estado = True #si el estado es cierto

for capital in capitales: #si una capital esta en la lista de capitales
    if texto == capital:#si el texto es una capital
        si_capital = True

if si_estado: #si es estado entonces imprime lo siguiente
    print("El texto corresponde a un ESTADO de la Repblica Mexicana.")
elif si_capital:#si es capital se imprime lo siguiente
    print("El texto corresponde a una CAPITAL de la Repblica Mexicana.")
else: #si no se detecta como estado o capital entonces imprime el texto agregado y una leyenda diciendo que no se encuentra en las listas
    print(texto, "no se encuentra como estado o capital de la Repblica Mexicana.")

```

```python
lista_de_numeros = [] #creamos una lista llamada numeros
suma = 0 #en esta variable se iniciara la suma, empezando con 0 e ira aumentando segun los numeros proporcionados por el usuario

while suma <= 100:#mientras que la suma de los numeros sea menor o igual a 100 debe de seguir
    num = float(input("Ingresa un nmero: ")) #agregamos una variable en la cual se iran introduciendo los nmeros por el usuario
    lista_de_numeros.append(num) #utilizamos el append para agregar num(los numeros proporcionados por el usuario) a la lista de numeros
    suma += num #a la suma se le estara agregando el numero proporcionado por el usuario, pero ya sumando

print("\nLista completa:", lista_de_numeros) #imprimimos la lista completa
print("Suma total:", suma) #mostramos la suma final sin rebasar 100
```

```python
datos = [] #creamos una lista vacia
continuar = True #mientras se continue el ciclo sigue

while continuar: #el programa se repetir mientra continuar sea True
    print("\n--- MEN ---") #mostramos el men de opciones
    print("1) Aadir nmero")
    print("2) Quitar ltimo nmero")
    print("3) Ver lista")
    print("4) Terminar")

    eleccion = input("Elige una opcin: ") #agregamos para que el usuario introduzca una opcion

    if eleccion == "1": #si la eleccion es 1, sigue
        valor = float(input("Escribe un nmero: ")) #introducimos un numero float
        datos.append(valor) #lo agregamos a la lista datos
        print("Nmero guardado.") #mostramos que el numero se ha guardado

    elif eleccion == "2": #si la eleccion en el menu es 2, continua
        if datos:
            quitado = datos.pop() #si hay datos, el ultimo se eliminara
            print("Nmero eliminado:", quitado) #mostramos que el numero se ha quitado
        else:
            print("No hay nmeros en la lista.") #si no hay numeros mostramos que no hay numeros en la lista

    elif eleccion == "3": #si la eleccion es 3 seguimos
        print("Lista actual:", datos) #mostramos la lista de datos

    elif eleccion == "4": # si la opcion es 4
        continuar = False #continuar cambia a estado False y el while se detiene
        print("Programa finalizado.") #mostranmos que el programa ha finalizado

    else:
        print("Opcin incorrecta.") #si no se introduce un numero del 1 al 4 mostramos opcion incorrecta y se repite hasta introducir una opcion valida

```
