Crear funcion que convierta temperatura Celsius a Kelvin

```python
def conversionC_K(T_C): #con esto ya quedo definida la función
  res = T_C + 273.15
  return res
```

```python
conversionC_K(25)
print('La temperatura es: ', conversionC_K(25), 'K')
```

Funcion para pasar de bar a MPa

```python
def convbar_Mpa(P_bar): #Definimos la función
  resultado = P_bar / 10 #Ponemos el factor de conversion de bar a MPa
  return resultado
```

```python
convbar_Mpa(15)
print('La presión es: ', convbar_Mpa(15), 'MPa')
```

Funcion para pasar las temperaturas a Kelvin

```python
def conversionT(T, unidad):
  if unidad == 'C':
    res = T + 273.15
    return res
  elif unidad == 'R':
    res = T * 5/9
    return res
```

```python
conversionT(25,'R')
print('La temperatura es: ',conversionT(25,'R'), 'K')
```

Ahora agregamos tambien para kelvin

```python
T = int(input('Ingresa la temperatura: '))
unidad = input('Ingresa la unidad: ')
conversionT(T,unidad)
def conversionT(T, unidad):
  if unidad == 'C':
    res = T + 273.15
    return res
  elif unidad == 'R':
    res = T * 5/9
    return res
  else:
    res = (T - 32) * 5/9 + 273.15 #agregamos para Farenheit
    return res

print('La temperatura es: ',conversionT(T,unidad), 'K')
```

```python
conversionT(25,'F')
print('La temperatura es: ',conversionT(25,'F'), 'K')
```

bar a MPa, psia a MPa, Kpa a MPa, atma a MPa

```python
def conversionP(P,unidad): #primero definimos la la funcion dando el nombre de conversionP, agregamos que la variable P, pero no tiene unidad
  if unidad == 'bar': #lo primero es cuando la unidad es bar
    res = P / 10
    return res #si esto se cumple que nos regrese res
  elif unidad == 'psia': #para psia
    res = P / 145
    return res
  elif unidad == 'Kpa': #para Kpa
    res = P / 1000
    return res
  else: #para atm
    res = P / 9.869
    return res
```

```python
conversionP(59,'Kpa') #llamamos la funcion con los valores y unidades que queramos
print('La presion es: ',conversionP(59,'Kpa'), 'MPa') #imprimimos con print para poder tener texto, tenemos que llamar la funcion
```

```python
def reynolds(Re):
  if Re < 2100:
    res = 'Laminar'
    return res
  elif Re < 4000 and Re > 2100:
    res = 'Transicion'
    return res
  else:
    res = 'Turbulento'
  return res
```
