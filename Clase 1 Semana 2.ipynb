Dada una ecuación cuadrática.

ax^2 + bx + c= 0,

El discriminante (Δ=b^2-4ac) determina el tipo de raíces:

Δ>0, dos raíces reales distintas

Δ = 0 una raíz real doble

Δ < 0 dos raíces complejas conjugadas

Ejercicio: Escribe un código que indique el tipo de raíces e imprima el valor si son reales.

```python
a = 34 #Asignamos valores directamente a a,b,c
b = 45
c = 57

delta = (b**2)-(4*a*c) #es la ecuacion de delta
print(f'El determinante es {delta:.2f}') #agregamos f  para que ns de valores con punto decimal y lo que esta dentro de {} es para que nos de 2 valores despues del punto decimal

if delta > 0:
    print('Hay dos raíces reales distintas')
elif delta == 0:
    print('Hay una raíz real doble')
else:
    print('Hay dos raíces complejas conjugadas')
```

```python
a = 5
b = 10
c = 2

delta = (b**2)-(4*a*c) #ecuación de delta
if delta > 0:
    x1 = (-b + (delta**0.5))/(2*a)
    x2 = (-b - (delta**0.5))/(2*a)
    print('Dos raíces reales distintas', x1, x2)
elif delta == 0:
    x = -b/(2*a)
    print('Una raíz real doble', x)
else:
  print('Raíces complejas conjugadas')
```

Dada una ecuación cuadrática.

ax^2 + bx + c= 0,

El discriminante (Δ=b^2-4ac) determina el tipo de raíces:

Δ>0, dos raíces reales distintas

Δ = 0 una raíz real doble

Δ < 0 dos raíces complejas conjugadas

Ejercicio: Escribe un código que indique el tipo de raíces e imprima el valor si son reales.

```python
a = .3
b = 2
c = 1000

delta = (b**2)-(4*a*c) #ecuación de delta
if delta > 0:
    x1 = (-b + math.sqrt(delta))/(2*a)
    x2 = (-b - math.sqrt(delta))/(2*a)
    print('Dos raíces reales distintas', x1, x2)
elif delta == 0:
    x = -b/(2*a)
    print('Una raíz real doble', x)
else:
  print('Raíces complejas conjugadas')
```

**Bucle while**

Se usa cuando no sabemos cuantas veces se repetira, solo sabemos que queremos repetir mientras se cumpla una condicion.

```python
contador = 1 #El bucle se repite mientras la condicion sea verdadera

while contador <= 5:
    print(contador)
    contador = contador + 1
```

Break, continue

```python
L = [1, 2, None, 4, None, 5]

suma = 0

for i in L:
  if i is None: continue
  suma += i

print(suma)
```
